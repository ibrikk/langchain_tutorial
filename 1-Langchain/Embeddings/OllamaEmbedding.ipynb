## Data Ingestion -> Data Transformation (dividing into chunks) -> Embedding (text to vectors) -> VectorStoreDB (FAISS, CMROMA DB, AstraDB)


## Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) apps that combine text promots with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = (
    OllamaEmbeddings(model="gemma:2b") ## by defualt it uses llama2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first character of the Greek alphabet",
        "Beta is the second character of the Greek alphabet",
    ]
)
```

```python
len(r1[0])
```

```python
embeddings.embed_query("What is the second letter of the Greek alphabet")
```

```python
### Other embedding models

embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document."
query_result = embeddings.embed_query(text)
len(query_result)
```
